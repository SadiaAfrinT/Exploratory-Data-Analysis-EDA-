#### Making one and n dimensional array by default functions

```python
"""one dimensional array"""
array_1 = np.arange(10) #create an array with 10 elements
print("Array is: {} \nand the type: {}\nlength: {}".format(array_1, type(array_1), len(array_1)))
```

```python
"""ndimensional array"""
array_2 = np.arange(10).reshape(5,2) #create 5x2 dimensional matrix
print("Array:\n {} \nand the type: {}\nlength: {}".format(array_2, type(array_2), len(array_2)))
print("Shape:",array_2.shape)
```

**Note:** we have some default matrix or array type, for example,

```python
"""zero matrix"""
a = np.zeros((2,2))
print("Array:\n {} \nand the type: {}\nlength: {}".format(a, type(a), len(a)))
print("Shape:",a.shape)
```

```python
"""one matrix"""
b = np.ones((2,2))
print("Array:\n {} \nand the type: {}\nlength: {}".format(b, type(b), len(b)))
print("Shape:",b.shape)
```

### More on Selecting Rows and Columns

```python
"""create a data frame consisting of the weight and origin of the subset of cars with year greater than 80"""
idx_80 = Auto_re['year'] > 80
Auto_re.loc[idx_80, ['weight', 'origin']]
```

```python
"""doing the same task by using the funtion lambda"""
Auto_re.loc[lambda df: df['year'] > 80, ['weight', 'origin']]
```

```python
""" all cars built after 1980 that achieve greater than 30 miles per gallon : using lambda function"""
Auto_re.loc[lambda df: (df['year'] > 80) & (df['mpg'] > 30),['weight', 'origin']]
```

```python
"""retrieve all Ford and Datsun cars with displacement less than 300. We check whether each name entry contains
either the string ford or datsun using the str.contains() method of the index attribute of of the dataframe"""
Auto_re.loc[lambda df: (df['displacement'] < 300) & (df.index.str.contains('ford')| df.index.str.contains('datsun')),
            ['weight', 'origin']]
```

**Take away message:** For integer based queries, use the iloc[] method. For string and Boolean selections, use the loc[] method. For functional queries that flter rows, use the loc[] method with a function (typically a lambda) in the rows argument.

## For Loop
- standard tool that repeatedly evaluates for some chunk of code while varying diferent values inside the code

```python
"""print only the last total sum"""
total = 0
for value in [3,2,19]:
    total += value
print('Total is: {0}'.format(total))
```

```python
"""Loops can be nested by additional indentation"""
total = 0
for value in [2,3,19]:
    for weight in [3, 2, 1]:
        total += value * weight
print('Total is: {0}'.format(total))
```

```python
"""To compute the average value of a random variable that takes on possible values 2, 3 or 19 
   with probability 0.2, 0.3, 0.5 respectively we would compute the weighted sum. 
   Tasks such as this can often be accomplished using the zip() function that loops over a sequence of tuples."""

total = 0
for value, weight in zip([2,3,19],[0.2,0.3,0.5]):
    total += weight * value
print('Weighted average is: {0}'.format(total))
```

```python
"""Creating a dataFrame"""
rng = np.random.default_rng(1) #create random variables
A = rng.standard_normal((127, 5)) #create random variables
M = rng.choice([0, np.nan], p=[0.8,0.2], size=A.shape) # random choice with selected probability
A += M #adding variables
D = pd.DataFrame(A, columns=['food',
'bar',
'pickle',
'snack',
'popcorn']) # create a Pandas DataFrame : D
```
